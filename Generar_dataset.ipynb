## AUMENTO DE CONJUNTO DE DATOS

```python
# Función para obtener sinónimos
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word, lang='spa'):
        for lemma in syn.lemmas('spa'):
            synonyms.add(lemma.name())
    return list(synonyms)

# Función para parafraseo avanzado
def parafraseo_avanzado(texto):
    doc = nlp(texto)
    palabras = []
    for token in doc:
        if token.pos_ in ['NOUN', 'VERB', 'ADJ']:
            sinonimos = get_synonyms(token.text)
            if sinonimos:
                palabras.append(random.choice(sinonimos))
            else:
                palabras.append(token.text)
        else:
            palabras.append(token.text)
    return ' '.join(palabras)

# Función para generar variaciones de preguntas
def generar_variacion(pregunta):
    prefijos = [
        "¿Podrías decirme ", "Me gustaría saber ", "¿Sabes ", "Necesito información sobre ",
        "¿Tienes datos acerca de ", "¿Podrías explicarme ", "Estoy interesado en conocer ",
        "¿Me ayudarías a entender ", "Quisiera que me aclares ", "¿Puedes informarme sobre "
    ]
    sufijos = [
        " por favor?", " si es posible.", ".", " ¿Puedes ayudarme?",
        " Te lo agradecería.", " Es importante para mí.", " ¿Qué me puedes decir al respecto?",
        " Necesito esta información.", " ¿Tienes datos sobre esto?", " Si no es mucha molestia."
    ]
    pregunta_base = pregunta.lower().rstrip('?')
    pregunta_parafraseada = parafraseo_avanzado(pregunta_base)
    return random.choice(prefijos) + pregunta_parafraseada + random.choice(sufijos)


# Generar variaciones y añadirlas al DataFrame
for _ in range(6):  # Genera 6 variaciones por cada pregunta original
    nuevas_filas = []
    for _, row in df.iterrows():
        nueva_pregunta = generar_variacion(row['pregunta'])
        nuevas_filas.append({
            'categoria': row['categoria'],
            'pregunta': nueva_pregunta,
            'respuesta': row['respuesta']
        })
    df = pd.concat([df, pd.DataFrame(nuevas_filas)], ignore_index=True)

# Mezclar el DataFrame
df = df.sample(frac=1).reset_index(drop=True)

# Guardar en CSV en Google Drive
ruta_guardado = '/content/drive/My Drive/LLM/datos_chatbot_soporte_tecnico.csv'
df.to_csv(ruta_guardado, index=False)
print(f"Se han generado y guardado {len(df)} ejemplos en '{ruta_guardado}'")

print("\nDistribución de categorías:")
print(df['categoria'].value_counts())

# Mostrar las primeras filas
print(df.head())
```
